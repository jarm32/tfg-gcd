```python
journal_set=set()

for csv_file_path in csvs:
    try:
        # Abrir y leer el archivo CSV
        with open(csv_file_path, mode="r", encoding="latin-1") as csv_file:
            csv_reader = csv.DictReader(csv_file)
            
            # Recorrer las filas del CSV
            for row in csv_reader:
                # Acceder a la columna "journal"
                if "journal" in row:
                    value = row["journal"]
                    # Filtrar valores None y cadenas vacias
                    if value not in (None, ''):
                        journal_set.add(value)
    
    except FileNotFoundError:
        print(f"El archivo '{csv_file_path}' no se encontro.")
    except KeyError:
        print("La columna 'journal' no se encontro en el archivo CSV.")
    except Exception as e:
        print(f"Ocurrio un error: {e}")
```

```python
def agrupar_palabras(lista):
    resultado = {}
    normalizados = defaultdict(list)

    # Agrupamos las palabras ignorando mayusculas/minusculas
    for palabra in lista:
        clave_normalizada = palabra.lower()
        normalizados[clave_normalizada].append(palabra)

    # Creamos el diccionario final con la primera ocurrencia como clave
    for clave_normalizada, palabras in normalizados.items():
        clave = palabras[0]  # Primera ocurrencia
        valores = palabras[1:]  # Resto de las ocurrencias
        resultado[clave] = valores

    return resultado
```

```python
# Vuelve a crear los csv con los cambios realizados

for csv_file_path in csvs:
    with open(csv_file_path, mode="r", encoding="latin-1") as csv_file:
        csv_reader = csv.DictReader(csv_file)
        
        # Crear una lista para guardar las filas modificadas
        filas_modificadas = []
        
        # Recorrer las filas del CSV
        for row in csv_reader:
            # Acceder a la columna "journal"
            if "journal" in row:
                value = row["journal"]
                if value is not None:
                    value = value.strip()
                
                for k,v in cambio.items():
                    if value in v:
                        # Cambiar el valor de "journal" a k
                        row["journal"] = k
            
            # Reemplazar valores None por espacios vacios (" ")
            # Asegurarnos de que no haya claves None en las filas
            row = {key: (value if value is not None else " ") for key, value in row.items()}
            
            # Asegurarnos de que no haya claves no validas
            for key in list(row.keys()):
                if key is None:
                    del row[key]

            # Agregar la fila modificada a la lista
            filas_modificadas.append(row)

    nombre = csv_file_path.split(".")
    nombre[0] = "0" + nombre[0] + "_arreglado"
    csv_file_path = ".".join(nombre)

    # Guardar el archivo CSV con los cambios
    with open(csv_file_path, mode="w", encoding="latin-1", newline="") as csv_file:
        # Obtener los encabezados (fieldnames)
        if filas_modificadas:
            fieldnames = filas_modificadas[0].keys()
        else:
            fieldnames = []  # En caso de que no haya filas modificadas

        # Crear el escritor de CSV con las mismas columnas
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        
        # Escribir los encabezados
        csv_writer.writeheader()
        
        # Escribir las filas modificadas
        csv_writer.writerows(filas_modificadas)
```
