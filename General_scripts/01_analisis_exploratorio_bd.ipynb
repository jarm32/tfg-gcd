```python
def formato_coma_decimal(x, pos):
    return f"{x:.2f}".replace(".", ",")

def formato_coma_decimal_1(x, pos):
    return f"{x:.1f}".replace(".", ",")
```

```python
def read_csv_auto(path, **kwargs):
    """
    Lee un CSV probando varios encodings y separadores habituales.
    Ãtil para ficheros separados por coma, punto y coma o tabulador.
    """
    encodings = ["utf-8", "utf-8-sig", "latin1", "cp1252"]
    seps = [",", ";", "\t"]
    last_error = None
    
    for enc in encodings:
        for sep in seps:
            try:
                df = pd.read_csv(
                    path,
                    encoding=enc,
                    sep=sep,
                    engine="python",
                    **kwargs
                )
                
                # Evita aceptar lecturas malas con todo metido en una sola columna
                if df.shape[1] > 1:
                    print(f"LeÃ­do correctamente: {path.name} | encoding={enc} | sep={repr(sep)}")
                    return df
                
            except Exception as e:
                last_error = e
    
    raise last_error


def split_list_cell(x):
    """
    Convierte celdas tipo 'a, b, c' o listas en texto en una lista limpia.
    """
    if pd.isna(x):
        return []
    
    if isinstance(x, list):
        return [str(v).strip() for v in x if str(v).strip()]
    
    x = str(x).strip()
    if x == "" or x.lower() in {"nan", "none", "null"}:
        return []
    
    # Limpieza bÃ¡sica si aparece como lista textual
    x = x.replace("[", "").replace("]", "").replace("'", "").replace('"', "")
    
    parts = [p.strip() for p in re.split(r",|;", x) if p.strip()]
    return parts


def clean_text(x):
    if pd.isna(x):
        return ""
    return str(x).strip()


def save_fig(filename):
    path = OUT_DIR / filename
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Guardado:", path)
```

```python
# Primera columna como identificador y resto como palabras clave
inv_kw_values = mat_inv_kw.iloc[:, 1:]
rev_kw_values = mat_rev_kw.iloc[:, 1:]

# Densidad: proporciÃ³n de unos respecto al total de celdas binarias
densidad_inv_kw = (inv_kw_values == 1).sum().sum() / inv_kw_values.size
densidad_rev_kw = (rev_kw_values == 1).sum().sum() / rev_kw_values.size

tabla_semantica = pd.DataFrame({
    "Elemento": [
        "Investigadores con representaciÃ³n temÃ¡tica",
        "Revistas con representaciÃ³n temÃ¡tica",
        "Palabras clave Ãºnicas en matriz de investigadores",
        "Palabras clave Ãºnicas en matriz de revistas",
        "Relaciones investigador-palabra clave activas",
        "Relaciones revista-palabra clave activas",
        "Densidad matriz investigador-palabra clave",
        "Densidad matriz revista-palabra clave"
    ],
    "Valor": [
        len(mat_inv_kw),
        len(mat_rev_kw),
        mat_inv_kw.shape[1] - 1,
        mat_rev_kw.shape[1] - 1,
        int((inv_kw_values == 1).sum().sum()),
        int((rev_kw_values == 1).sum().sum()),
        round(densidad_inv_kw, 4),
        round(densidad_rev_kw, 4)
    ]
})

display(tabla_semantica)

tabla_semantica.to_csv(OUT_DIR / "eda_tabla_cobertura_semantica.csv", index=False, encoding="utf-8-sig")
print("Guardado:", OUT_DIR / "eda_tabla_cobertura_semantica.csv")
```

```python
# ========== PROCESAMIENTO COMPLETO DE PROYECTOS ==========

import warnings
warnings.filterwarnings('ignore')

# 1. Leer todos los archivos de proyectos
carpeta_proyectos = "C:/Code/tfg-gcd/__Interface"
archivos_proyectos = glob.glob(os.path.join(carpeta_proyectos, "ha_participado_en_*_FINAL.xlsx"))

print("="*80)
print("PROCESAMIENTO DE PROYECTOS")
print("="*80)

dfs_proyectos = []
num_archivos = 0

for archivo in archivos_proyectos:
    try:
        df = pd.read_excel(archivo)
        dfs_proyectos.append(df)
        num_archivos += 1
    except Exception as e:
        print(f"Error al leer {os.path.basename(archivo)}: {e}")

df_proyectos_total = pd.concat(dfs_proyectos, ignore_index=True)

print(f"\n1. NÃºmero de archivos leÃ­dos: {num_archivos}")
print(f"   Archivos encontrados: {len(archivos_proyectos)}")

# 2. Detectar columnas
print(f"\n2. Columnas detectadas:")
columnas = list(df_proyectos_total.columns)
for col in columnas:
    print(f"   - {col}")

# 3. InformaciÃ³n del dataset completo
print(f"\n3. InformaciÃ³n del dataset:")
print(f"   NÃºmero total de registros: {len(df_proyectos_total)}")

# 4. Extraer aÃ±o de la columna FECHA
df_proyectos_total['AÃO'] = pd.to_datetime(
    df_proyectos_total['FECHA'], 
    format='%Y/%m/%d',
    errors='coerce'
).dt.year

# Verificar si hay fechas invÃ¡lidas
num_fechas_invalidas = df_proyectos_total['AÃO'].isna().sum()
if num_fechas_invalidas > 0:
    print(f"   â  Advertencia: {num_fechas_invalidas} registros con fecha invÃ¡lida")

# 5. Crear clave Ãºnica de proyecto usando solo el tÃ­tulo normalizado
df_proyectos_total['PROYECTO_UNICO'] = (
    df_proyectos_total['TÃTULO']
    .astype(str)
    .str.strip()
    .str.lower()
)

# Para cada proyecto Ãºnico, conservar el primer aÃ±o disponible
df_proyectos_unicos = (
    df_proyectos_total
    .dropna(subset=['AÃO'])
    .sort_values('AÃO')
    .drop_duplicates(subset=['PROYECTO_UNICO'], keep='first')
)

print(f"\n4. Proyectos Ãºnicos:")
print(f"   Registros originales: {len(df_proyectos_total)}")
print(f"   Registros Ãºnicos (TÃTULO): {len(df_proyectos_unicos)}")

# 6. Contar proyectos por aÃ±o
conteo_por_anio = df_proyectos_unicos['AÃO'].value_counts().sort_index()

print(f"\n5. DistribuciÃ³n de proyectos por aÃ±o:")
print(conteo_por_anio.to_string())

print("\n" + "="*80)
```
