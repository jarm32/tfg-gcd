```python
import requests
import csv

# ORCID iD del investigador
orcid_id = '0000-0001-5564-5365'

# URL de la API pública de ORCID para obtener las actividades del investigador
url = f'https://pub.orcid.org/v3.0/{orcid_id}/works'

# Encabezados para la solicitud HTTP
headers = {
    'Accept': 'application/json'
}

# Realizar la solicitud GET a la API de ORCID
response = requests.get(url, headers=headers)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    # Analizar la respuesta JSON
    data = response.json()
    works = data.get('group', [])

    # Lista para almacenar los datos extraídos
    extracted_data = []

    # Recorrer cada trabajo y extraer la información deseada
    for work in works:
        # Título del artículo
        title = work.get('work-summary', [{}])[0].get('title', {}).get('title', {}).get('value', 'No disponible')

        # Año de publicación
        publication_date = work.get('work-summary', [{}])[0].get('publication-date', {})
        year = publication_date.get('year', {}).get('value', 'No disponible')

        # Nombre de la revista
        journal_title = work.get('work-summary', [{}])[0].get('journal-title', {}).get('value', 'No disponible')

        # Contribuidores (autores)
        contributors = work.get('work-summary', [{}])[0].get('contributors', {}).get('contributor', [])
        contributor_names = ', '.join([contributor.get('credit-name', {}).get('value', 'No disponible') for contributor in contributors])

        # Agregar los datos a la lista
        extracted_data.append([title, journal_title, year, contributor_names])

    # Escribir los datos en un archivo CSV
    with open('datos_finales_Orcid_ALV.csv', 'w', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        # Escribir los encabezados
        csvwriter.writerow(['title', 'journal', 'year', 'otros_investigadores'])
        # Escribir las filas de datos
        csvwriter.writerows(extracted_data)

    print("Datos extraídos y guardados en datos_finales_Orcid_ALV.csv exitosamente.")
else:
    print(f"Error al acceder a la API de ORCID: {response.status_code}")

```
