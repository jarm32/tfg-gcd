```python
import requests
from bs4 import BeautifulSoup
import unidecode

url = 'https://iumpa.webs.upv.es/grupos-y-autores/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    nombres = soup.find_all('h3', class_='et_pb_jv_team_title')

    lista_nombres_reves = []
    for nombre in nombres:
        if nombre != 'Murillo Arcila, Marina':
            lista_nombres_reves.append(nombre.get_text(strip=True))
        
    

    print(lista_nombres_reves)
else:
    print(f"Error al acceder a la página: {response.status_code}")
```

```python
lista_nombres_reformateados = []

for nombre in lista_nombres_reves:
    if ',' in nombre:
        apellidos, nombre_propio = nombre.split(',', 1)
        nombre_reformateado = f"{nombre_propio.strip()} {apellidos.strip()}"
        lista_nombres_reformateados.append(nombre_reformateado)
    else:
        lista_nombres_reformateados.append(nombre)

print(lista_nombres_reformateados)
```

```python
def formatear_nombres_bien(nombres):
    nombres_formateados = []
    for nombre in nombres:
        nombre_formateado = unidecode.unidecode(nombre.lower()).replace(' ', '-')
        nombres_formateados.append(nombre_formateado)
    return nombres_formateados

nombres_guion_bien = formatear_nombres_bien(lista_nombres_reformateados)
nombres_guion_bien
```

```python
def formatear_nombres_reves(nombres):
    nombres_formateados = []
    for nombre in nombres:
        nombre_sin_coma = nombre.replace(',', '')
        nombre_formateado = unidecode.unidecode(nombre_sin_coma.lower()).replace(' ', '-')
        nombres_formateados.append(nombre_formateado)
    return nombres_formateados

nombres_guion_reves = formatear_nombres_reves(lista_nombres_reves)
nombres_guion_reves
```

```python
def obtener_correo(url_nombre, guion_reves, index):
    url = f'https://iumpa.webs.upv.es/miembros/{url_nombre}/'

    try:
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            correo_tag = soup.find('div', class_='et_pb_blurb_description')
            correo = correo_tag.get_text(strip=True) if correo_tag else "No tiene correo asociado"

            return correo
        else:
            if index < len(guion_reves):
                nuevo_nombre = guion_reves[index]
                nueva_url = f'https://iumpa.webs.upv.es/miembros/{nuevo_nombre}/'
                response = requests.get(nueva_url)
                if response.status_code == 200:
                    soup = BeautifulSoup(response.text, 'html.parser')
                    correo_tag = soup.find('div', class_='et_pb_blurb_description')
                    correo = correo_tag.get_text(strip=True) if correo_tag else "No tiene correo asociado"
                    return correo

            return "Error al acceder a la página"
    except Exception as e:
        return "Error durante la extracción"
correos = [obtener_correo(elem, nombres_guion_reves, index) for index, elem in enumerate(nombres_guion_bien)]
```
