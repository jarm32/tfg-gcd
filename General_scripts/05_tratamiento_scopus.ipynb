```python
import pandas as pd

# Cargar el archivo CSV
input_file = "datos_Scop_SLJM.csv"
output_file = "datos_finales_Scop_SLJM.csv"

# Leer el archivo CSV
df = pd.read_csv(input_file)

# Seleccionar y renombrar las columnas
df_nuevo = df[["Title", "Publisher", "Year", "Authors"]].rename(columns={
    "Title": "title",
    "Publisher": "journal",
    "Year": "year",
    "Authors": "otros_investigadores"
})

# Guardar el nuevo archivo CSV
df_nuevo.to_csv(output_file, index=False, encoding="utf-8")

print(f"Archivo guardado en: {output_file}")
```
