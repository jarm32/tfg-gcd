```python
import pandas as pd
from unidecode import unidecode

def convertir_columna_ascii(df, columna):
    if columna not in df.columns:
        raise ValueError(f"La columna '{columna}' no existe en el DataFrame.")
    
    df[columna] = df[columna].astype(str).apply(unidecode)
    
    return df
```

```python
df = pd.read_csv('Investigadores_internos.csv', encoding='latin1')
```

```python
ndf = convertir_columna_ascii(df, "Nombre")
```

```python
df.to_csv('Investigadores_internos.csv', index=False)
```
